```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\Karishma\Data analysis classes\projects\Supply Chain & Inventory Optimization Analysis\Supply Chain proect\supply_chain_raw_50k.csv")
df
```

```python
print("Before Cleaning:")
print(df.info())
```

```python
# Fix Date Column

df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
```

```python
#Convert Numeric Columns

numeric_cols = ['stock_quantity','order_quantity','cost_price','selling_price','delivery_days']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
#Handle Missing Values

df['stock_quantity'] = df['stock_quantity'].fillna(df['stock_quantity'].median())
df['order_quantity'] = df['order_quantity'].fillna(0)
df['delivery_days'] = df['delivery_days'].fillna(df['delivery_days'].median())
```

```python
#Remove Negative Stock

df.loc[df['stock_quantity'] < 0, 'stock_quantity'] = 0
```

```python
#Drop Invalid Dates

df = df.dropna(subset=['order_date'])

```

```python
#Feature Engineering

df['Revenue'] = df['order_quantity'] * df['selling_price']
df['Cost'] = df['order_quantity'] * df['cost_price']
df['Profit'] = df['Revenue'] - df['Cost']

df['Inventory_Turnover'] = df['order_quantity'] / df['stock_quantity']
df['Stockout'] = np.where(df['stock_quantity'] == 0, 1, 0)
```

```python
print("After Cleaning:")
print(df.info())

df.to_csv("D:\Karishma\Data analysis classes\projects\Supply Chain & Inventory Optimization Analysis\Supply Chain proect\supply_chain_cleaned.csv", index=False)
```
